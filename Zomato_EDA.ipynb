```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
final_df
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
Country_name=final_df.Country.value_counts().index
```

```python
Country_value=final_df.Country.value_counts().values
```

```python
## Top 3 countries that uses zomato
plt.pie(Country_value[:3],labels=Country_name[:3],autopct='%0.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

```python
##countplot
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
